```python
import altair as alt
from vega_datasets import data
from altair.expr import expr
import pandas as pd
```

Завдання 1: Гістограма типів погоди в Сіетлі

```python
weather = data.seattle_weather()

chart1 = alt.Chart(weather).mark_bar(cornerRadiusTopLeft=5, cornerRadiusTopRight=5).encode(
    x=alt.X('weather:N', sort='-y', title='Тип погоди'),
    y=alt.Y('count():Q', title='Кількість днів'),
    color='weather:N'
).properties(title='Кількість днів кожного типу погоди в Сіетлі')

chart1
```

Завдання 2: Горизонтальна стовпчикова діаграма врожайності

```python
barley = data.barley()

chart2 = alt.Chart(barley[barley['year'] == 1931]).mark_bar().encode(
    y=alt.Y('variety:N', sort='-x', title='Сорт'),
    x=alt.X('yield:Q', title='Врожайність'),
    color='site:N'
).properties(title='Врожайність сортів ячменю у 1931 році')

chart2
```

Завдання 3: Гістограма погоди по місяцях

```python
weather = data.seattle_weather()
weather['month'] = pd.to_datetime(weather['date']).dt.month_name()

chart3 = alt.Chart(weather).mark_bar().encode(
    x=alt.X('month:N', sort=list(pd.date_range('2000-01-01', periods=12, freq='M').strftime('%B')), title='Місяць'),
    y=alt.Y('count():Q', title='Кількість днів'),
    color='weather:N',
    column=alt.Column('weather:N', title='Тип погоди')
).properties(title='Кількість днів певного типу погоди по місяцях')

chart3
```

Завдання 4: Лінійно-обласна діаграма цін акцій

```python
stock_data = data.stocks()

subset = stock_data[stock_data['symbol'].isin(['GOOG', 'AAPL'])]

chart4 = alt.Chart(subset).mark_area(
    interpolate='monotone',
    line=True,
    opacity=0.8,
    color=alt.Gradient(
        gradient='linear',
        stops=[
            alt.GradientStop(color='lightgray', offset=0),
            alt.GradientStop(color='darkblue', offset=1)
        ],
        x1=0, x2=0, y1=1, y2=0
    )
).encode(
    x=alt.X('date:T', title='Дата'),
    y=alt.Y('price:Q', title='Ціна'),
    color='symbol:N'
).properties(
    title='Динаміка цін GOOG та AAPL із градієнтною областю',
    width=650,
    height=350
)

chart4
```

Завдання 5: Потоковий граф безробіття

```python
data_url = data.unemployment_across_industries.url

chart5 = alt.Chart(data_url).mark_area(
    interpolate='basis',
    opacity=0.85
).encode(
    x=alt.X('yearmonth(date):T', axis=alt.Axis(format='%Y', title='Рік')),
    y=alt.Y('sum(count):Q', stack='center', axis=None, title=''),
    color=alt.Color('series:N', scale=alt.Scale(scheme='category20c'), title='Галузь')
).interactive().properties(
    width=680,
    height=380,
    title="Динаміка безробіття по секторах США"
)

chart5
```

Завдання 6: Кільцева діаграма типів погоди

```python
weather = data.seattle_weather()

weather_counts = weather.groupby('weather').size().reset_index(name='count')
weather_counts['percentage'] = 100 * weather_counts['count'] / weather_counts['count'].sum()
avg_temp = weather['temp_max'].mean().round(1)

chart6 = alt.Chart(weather_counts).mark_arc(innerRadius=50).encode(
    theta='count:Q',
    color='weather:N',
    tooltip=['weather:N', 'percentage:Q']
).properties(title='Типи погоди у Сіетлі')

text = alt.Chart(pd.DataFrame({'text': [f'Сер. темп: {avg_temp}°C']})).mark_text(size=20).encode(
    text='text:N'
)

chart6 + text
```

Висновок: Гістограми розкривають перевагу сонячної погоди та сезонні зміни, а також порівнюють врожайність ячменю. Лінійно-обласна діаграма показує зміну цін акцій GOOG і AAPL, а потоковий графік ілюструє динаміку безробіття по галузях. Кільцева діаграма доповнює аналіз, підтверджуючи частку сонячної погоди. Усі графіки разом дозволяють побачити ключові закономірності та взаємозв’язки в даних.
