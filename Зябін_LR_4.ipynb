# **2.Візуалізація даних**

# Лінійний графік (відвідуваність Avila Adobe з часом)

```python
plt.figure(figsize=(16, 9))
plt.plot(df['month'], df['avila_adobe'], marker='o', linestyle='-', color='teal')
plt.title('Відвідуваність Avila Adobe з часом')
plt.xlabel('Дата')
plt.ylabel('Кількість відвідувачів')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Boxplot (розподіл відвідувачів Chinese American Museum по роках)

**Додаємо колонку 'year' з року**

```python
df['year'] = df['month'].dt.year
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='year', y='chinese_american_museum', data=df, palette='pastel')

plt.title('Розподіл відвідувачів Chinese American Museum по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість відвідувачів')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Heatmap (кореляція між музеями)

```python
museum_data = df.select_dtypes(include='number')

plt.figure(figsize=(12, 8))
sns.heatmap(museum_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Кореляція між музеями')
plt.tight_layout()
plt.show()
```

# Histogram (розподіл кількості відвідувачів Gateway to Nature Center)

```python
plt.figure(figsize=(10, 6))
plt.hist(df['gateway_to_nature_center'].dropna(), bins=10, color='lightgreen', edgecolor='black')

plt.title('Розподіл кількості відвідувачів Gateway to Nature Center')
plt.xlabel('Кількість відвідувачів')
plt.ylabel('Кількість місяців')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Barplot (середня кількість відвідувачів по музеях за всі роки)

```python
museum_means = df.select_dtypes(include='number').drop(columns='year').mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=museum_means.values, y=museum_means.index, palette='viridis')

plt.title('Середня кількість відвідувачів по музеях (усі роки)')
plt.xlabel('Середня кількість відвідувачів')
plt.ylabel('Музей')
plt.tight_layout()
plt.show()
```
