```python
import plotly. express as px
```

```python
df = px.data.iris() # iris is a pandas DataFrame
df
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size="petal_length", hover_data=['petal_width'],
                 title="Sepal: Width vs Length",
                 color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_layout(template="plotly_dark", width=1000, height=800)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="petal_length",  # градієнт за числовим значенням
    color_continuous_scale="Viridis",  # палітра градієнта
    title="Sepal Dimensions Colored by Petal Length"
)

fig.update_layout(template="plotly_dark", width=1000, height=700)
fig.show()

```

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    symbol="species",
    color_discrete_sequence=px.colors.qualitative.Prism,
    title="Sepal Dimensions by Species"
)
fig.update_layout(template="plotly_dark")
fig.show()
```

```python
df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", trendline="ols",template="plotly_dark")
fig. show()
```

```python
df = px.data.gapminder().query("continent == 'Oceania'")
print(df.head())
```

```python
fig = px.line(df, x='lifeExp', y='gdpPercap', color='country', template="plotly_dark")
fig.show()
```

```python
df = px.data.gapminder().query("year == 2007")

fig = px.scatter(
    df, x="gdpPercap", y="lifeExp",
    size="pop", color="continent",
    template="plotly_dark",
    hover_name="country", log_x=True,
    title="GDP vs Life Expectancy (2007)",
    size_max=60
)
fig.show()
```

```python
df = px.data.gapminder().query("country == 'Japan'")

fig = px.line(
    df, x="year", y="lifeExp",
    template="plotly_dark",
    title="Life Expectancy in Japan Over Time"
)
fig.show()
```

```python
df = px.data.iris()

fig = px.violin(
    df, y="petal_length",x="species",
    color="species", box=True, points="all",
    template="plotly_dark",
    title="Distribution of Petal Length by Species"
)
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

N = 1000
t = np.linspace(0, 5*np.pi, 100)
y = 2*np.cos(t)*(t-np.pi/4)

fig = go.Figure(data=go.Scatter(x=t, y=y, mode='markers'))

fig.update_layout(template="plotly_dark")

fig.show()
```

Загальний висновок:
Plotly є потужним інструментом для створення інтерактивних графіків, який забезпечує велику гнучкість у налаштуванні візуалізації даних, надаючи безліч варіантів тем, кольорів, стилів та інтерактивних елементів.

NumPy є основним інструментом для числових обчислень, що дозволяє створювати та маніпулювати даними для графіків.

Спільне використання цих бібліотек дає можливість ефективно візуалізувати складні дані, роблячи графіки інтерактивними та наочними.
